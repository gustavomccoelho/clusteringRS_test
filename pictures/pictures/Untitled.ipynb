```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f1 = pd.read_csv('f1_list_Feature_Recursive_Elimination_RandomForest.csv', index_col = 0, squeeze = True)
accuracy = pd.read_csv('accuracy_list_Feature_Recursive_Elimination_RandomForest.csv', index_col = 0, squeeze = True)
precision = pd.read_csv('precision_list_Feature_Recursive_Elimination_RandomForest.csv', index_col = 0, squeeze = True)
recall = pd.read_csv('recall_list_Feature_Recursive_Elimination_RandomForest.csv', index_col = 0, squeeze = True)
```

```python
plt.figure(figsize = (20, 10))
plt.plot(f1)
plt.plot(accuracy)
plt.plot(precision)
plt.plot(recall)
plt.legend(["F1", "Accuracy", "precision", "Recall"])
plt.xticks(rotation=80)
plt.show()
```
